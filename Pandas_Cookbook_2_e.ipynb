#1. Pandas 기초


*   DataFrame(2차원)에서 column을 선택해서 Series(1차원 데이터 셋)으로 반환


*   dataFrame: index, columns, data





```python
# 1) pandas 임포트
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/movie.csv"
movies = pd.read_csv(url)
print(movies.head())
print(movies.tail())
```



*   dataFrame은 column의 type이 다를 수 있으나 numpy 배열은 아님



```python
columns = movies.columns
index = movies.index
type(index)
```

```python
data = movies.to_numpy()
type(data)
```



*   Index(행 인덱스)와 Column(열 인덱스)은 밀접한 관계임, 둘다 Index의 서브클래스이므로, 서로 유사한 작업을 수행함


```python
issubclass(pd.RangeIndex, pd.Index)
```

```python
issubclass(columns.__class__, pd.Index)
```



*   작동원리: Index를 지정하지 않으면, pandas는 RangeIndex(Index의 하위 클래스) 사용 -> range객체와 유사함


*   **전체 값이 필요할 때까지 메모리에 로드되지 않아 메모리 절약됨**





```python
index.to_numpy() # hash table을 사용해 구현함
```

```python
columns.to_numpy()
```

### 데이터 형식 이해
*   Int64: NULL 값을 지원하는 pandas 정수 형식
*   object: 문자열(과 혼합형식)을 저장하는 numpy 형식
*   bool(결측치 x), boolean(NULL o)
*   datetime64[ns]: 날짜형식, 결측치(NaT) okay

```python
movies.dtypes       # 열의 이름, 데이터 형식 표기
movies.dtypes.value_counts() # 데이터 형식의 개수 반환
# movies.info()
```

```python
pd.Series(["Paul",np.nan, "George"]).dtype
# pandas는 문자열 + NaN이 섞여 있으면,이 경우 Python 객체를 담는 object 타입(series에서는 0)으로 표기
```

### 열 선택(Series 추출)


```python
#   .loc: 레이블 기반
movies.loc[:, "director_name"].head()
```

```python
#   .iloc: 위치 기반
movies.iloc[:, 1].head()
```
