### Series 연산

- 다양한 연산자를 다른 Series 객체에 적용해 완전히 다른 값을 가진 새 Series(pandas의 Series = 인덱스를 가진 1차원 배열)를 생성한다!

```python
url = "https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/movie.csv"
movies = pd.read_csv(url)

# 1. imdb_score 열을 Series로 선택한다
imdb_score = movies["imdb_score"]
imdb_score.head()
```

```python
# 2. 더하기 연산자 사용
(imdb_score + 1).head()

## .add 메서드를 사용해서 재현 가능함 → imdb_score.add(1)
pd.concat([imdb_score + 1, imdb_score.add(1)], axis=1)  # cbind (axis = 1 )
```

```python
# 3. 그 외 다른 산술연산자도 스칼라 값과 유사하게 작동함
(imdb_score * 2.5).head()
```

```python
# 4. 파이썬 //는 몫을 구하는 나눗셈 연산이며, 결과는 내림함. %는 나머지를 구하기
(imdb_score //7).head()
```

```python
# 5. 비교연산자( <, <=, >, >=, ==, != )도 존재하고, True, False로 반환함. 결과는 Boolean 배열임
imdb_score > 7

imdb_score.gt(7)
```

체인의 단점은 Debugging이 어려움.

Why? 생성 중에 생성된 중간 객체는 별도의 변수에 저장되지 않으므로 오류의 위치를 추적하기 어려움

→ 메서드 호출(method call) 을 자기 줄에 하나씩 둔 형태로 주석처리 하면서 작동여부를 확인하기


    fb_likes.fillna(0)
    # .astype(int)
    # .head()

    fb_likes.fillna(0).astype(int)
    # .head()


→ .pipe 매서드를 호출하여 중간 값을 표시하는 방안

```python
def debug_ser(ser):
  print("BEFOR")
  print(ser)
  print("AFTER")
  return ser

fb_likes.fillna(0).pipe(debug_ser).astype(int).head()
```

```python
intermediate = None
def get_intermediate(ser):
  global intermediate
  intermediate = ser
  return ser  # 전역변수 활용

res = (
    fb_likes.fillna(0)
    .pipe(get_intermediate)
    .astype(int)
    .head()
)

intermediate
```

```python
# 백슬래시를 이용해서 여러줄에 걸쳐 작성가능함(괄호로 묶어서 쓸 수도 있음)
fb_likes.fillna(0) \
.astype(int) \
.head()
```

```python
# 1. import data set
url = "https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/movie.csv"
movies = pd.read_csv(url)

# 2. rename DataFrame 메서드는 old : new colname을 매칭하는 Dic를 취함
col_map = {
    "director_name" : "director",
    "num_critic_for_reviews" : "critic_reviews"
}

# 3. Dictionary를 .rename 메서드에 전달하고, 결과를 새로운 변수에 할당함
movies.rename(columns = col_map).head()
```
