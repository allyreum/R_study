# **K-NN을 이용한 알고리즘**

````python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight # 리스트 합치는 방법
#print(length)
#print(weight)
````

**사이킷 런(머신러닝 패키지)을 이용하기 위해선 2차원 리스트 형태의 데이터가 필요함**

````python
fish_data = [[l,w] for l, w in zip(length, weight)]
# zip 함수는 나열된 리스트 각각 하나씩 원소를 꺼내 반환

print(fish_data[:5])
````

````python
fish_target = [1] * 35 + [0] * 14 # 결과값 생성: 1 = 도미, 0 = 빙어

print(fish_target)
````

K-NN 알고리즘을 구현한 클래스인 **KNeighborsClassifier** import

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
````

````python
kn.fit(fish_data, fish_target)      # trainning -> fit(): 주어진 데이터로 알고리즘 훈련
kn.score(fish_data, fish_target)    # model 평가(Accuracy)
````

(30, 600)의 생선을 분류하면?

````python
kn.predict([[30,600]]) # predict은 새로운 데이터의 정답을 예측
````

K-NN

장점: 새로운 데이터 예측 시엔 가장 가까운 직선거리의 데이터를 보게됨.

단점: 데이터 양이 큰 경우, 직선거리 산출 시 많은 시간 소요되고, 메모리로 많이 필요함.

````python
print(kn._fit_X)    # fish data를 모두 갖고 있음
print(kn._y)
````

**KNeighborsClassifier클래스는 기본적으로 가까운 5개의 데이터를 참고한다. (K = 5)**
하지만 **n_neighbors** 매개변수를 통해 변경이 가능하다.

````python
kn49 = KNeighborsClassifier(n_neighbors = 49)   # n_neighbors를 통해 매개변수 변경가능, k = 49개로 한 모델

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
````

accuracy가 1에서 0.71로 낮아진 걸로 보았을 때 k=5로 보는 것이 도미와 빙어를 분류하는데 좋은 모델이었던 것으로 보인다!

fish data는 도미가 35개, 빙어가 14개로 구성되어 있음.

kn49는 49개 중 도미 35개로 다수를 차지 하니까 어떤 데이터를 넣어도 무조건 도미를 예측할 것임.

````python
print(35/49)    # 정확도 = 맞춘 개수 / 전체 개수, kn49는 도미만 올바르게 맞치기 때문에 score() 매서드와 같은 값을 얻을 수 있음.
````

# 확인문제
5 <= k <= 49 중에 score < 1로 되는 k를 구하여라

````python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
  kn.n_neighbors = n # 인접 이웃 개수 설정
  score = kn.score(fish_data, fish_target) # 점수 개산
  if score < 1:
    print(n, score)
    break
````

sklearn.metrics 패키지 아래에 mean_absolute_error는 타깃과 예측의 절대갓 오차를 평균하여 반환

- 모델을 복잡하게 만드는 방법은 K를 줄이면 됨. k를 감소시키면, 국지적인 패턴에 민감해짐. k가 높을수록 일반적인 패턴을 따르게 됨
- 과대적합일 땐, 모델을 간단하게 만들면 됨(k를 늘림)



````python
knr.n_neighbors = 3   # k를 지정해주기만 하면 됨
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
print(knr.score(test_input, test_target))
````

## 3-2. 선형회귀

````python
import numpy as np
# insert data(perch_length, perch_weight)

from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state = 42)

train_input = train_input.reshape(-1,1) # sklearn에서 2차원 배열 만들어야함!!!
test_input = test_input.reshape(-1,1)

from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors = 3)
knr.fit(train_input, train_target)

#50cm인 농어의 무게 예측
print(knr.predict([[50]]))    # knr.predict([[50]]) (2차원 표기)
````

실제로 더 많은 무게가 나간다고 했을 때, 예측이 잘못된 이유는?

````python
import matplotlib.pyplot as plt
distances, indexes = knr.kneighbors([[50]]) # 50cm 농어의 이웃
# distances: 50cm 농어와 가장 가까운 이웃들과의 거리 (유클리드 거리)
# indexes: 그 이웃들이 훈련 데이터에서 몇 번째 샘플이었는지 알려주는 인덱스

plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker = 'D')
plt.scatter(50, knr.predict([[50]]), marker='^')  # 예측값 표시
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

````python
print(np.mean(train_target[indexes]))
print(knr.predict([[100]]))
print(knr.predict([[50]]))
````

50cm, 100cm 농어의 무게가 동일함

````python
distances, indexes = knr.kneighbors([[100]])
plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker = 'D')
plt.scatter(100, knr.predict([[100]]), marker='^')  # 예측값 표시
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

근접한 이웃을 기반으로 하기 때문에 농어의 길이가 길어도 무게는 늘어나지 않음.

따라서 **회귀**를 이용해서 구해야하 함!!!

**4) 로지스틱 회귀**

$Z\ \left(확률\right)=\ \beta _0+\beta _1X_1+\beta _2X_2+...+\beta _nX_n$

0 <= prob <= 1 값, 즉 z가 아주 큰 음수일 때 0이 되고, z가 아주 큰 양수일 때 1이되도록 바꿔야 함

이때, 시그모이드 함수(or 로지스틱 함수)를 사용하면 가능해짐 [link text](https://ko.wikipedia.org/wiki/%EB%A1%9C%EC%A7%80%EC%8A%A4%ED%8B%B1_%ED%9A%8C%EA%B7%80)

````python
import numpy as np
import matplotlib.pyplot as plt
z = np.arange(-5,5,0.1)
phi = 1 / (1 + np.exp(-z))  # 지수함수 = np.exp()
plt.plot(z, phi)
plt.xlabel('z')
plt.ylabel('phi')
plt.show()
````

numpy에서는 불리언 인덱싱(T/F)으로 행 선택 가능

*4-2) 다중 분류 수행하기*

max_iter 매개변수에서 반복 횟수 지정(default = 100)

이때 매개변수는 C이고, alpha와 반대로 작을 수록 규제가 커짐.(C의 default = 1)

````python
lr = LogisticRegression(C=20, max_iter =1000)
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target), lr.score(test_scaled, test_target))

print(lr.predict(test_scaled[:5]))

proba = lr.predict_proba(test_scaled[:5])
print(np.round(proba, decimals = 3))
````

## **4-2. 확률적 경사 하강법**

````python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
````

````python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state = 42)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
````

**확률적 경사 하강법 대표 class는 SGDClassifier**

*2개의 매개변수 지정, max_iter는 수행할 에포크, 손실 함수의 종류를 지정(loss = log_loss, 로지스틱 손실함수)*

````python
from sklearn.linear_model import SGDClassifier
sc = SGDClassifier(loss = 'log_loss', max_iter = 10, random_state = 42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

*객체를 만들지 않고, 훈현한 모델 sc를 추가 훈련을 한다?  partial_fit()*

*SGDClassifier는 1epoch씩 이어서 훈련가능(미니배치,배치 하강법이 아님!)*

````python
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

***에포크와 과대/과소적합***

조기종료(early stopping): 과대적합이 시작하기 전에 훈련을 종료

````python
import numpy as np
sc = SGDClassifier(loss = 'log_loss', random_state = 42)
train_score = []
test_score = []
classes = np.unique(train_target) # target의 class 목록

for _ in range(0,300):    # _는 사용하지 않고 그냥 버리는 값을 넣어주는 용도(여기선 0)
  sc.partial_fit(train_scaled, train_target, classes = classes) # 전체 sample의 class 목록을 전달해야함
  train_score.append(sc.score(train_scaled, train_target))
  test_score.append(sc.score(test_scaled, test_target))

import matplotlib.pyplot as plt
plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show() # 반복횟수는 100가 적합해 보임
````

*SGDClassifier의 반복횟수를 100에 맞춰서 다시 훈련*
*   에포크동안 성능이 향상되지 않으면 더 훈련하지 않고 멈춤
*   tol 매개변수에서는 향상될 최솟값을 설정 -> None으로 지정하면, 멈추지 않고 100(max_iter)까지 무조건 반복

````python
sc = SGDClassifier(loss = 'log_loss', max_iter = 100, tol = None, random_state = 42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

*loss 매개변수의 기본값은 hinge. hinge loss는 SVM 알고리즘을 위한 손실함수*

````python
sc = SGDClassifier(loss = 'hinge', max_iter = 100, tol = None, random_state = 42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
````

##5-2. 교차 검증과 그리드 서치
**학습목적: 검증 세트가 필요한 이유, 그리드서치& 랜덤서치를 이용해 최적의 성능을 내는 하이퍼파라미터 찾기**

- 전체 데이터 구성: 훈련세트 60%, 검증세트 20%, 테스트세트 20%
(보통 2~30%를 테스트 세트와 검증 세트를 떼어 놓음)

````python
# import data
import pandas as pd
wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[['alcohol','sugar','pH']].to_numpy()
target = wine['class'].to_numpy()

# split data
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size = 0.2, random_state = 42
)

# split validation data
sub_input, val_input, sub_target, val_target = train_test_split(
    train_input, train_target, test_size = 0.2, random_state = 42
) # test_size = 0.2 : train_input의 약 20%를 val_input으로 만듦

print(sub_input.shape, val_input.shape)
````

````python
# DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 42)
dt.fit(sub_input, sub_target)
print(dt.score(sub_input, sub_target))
print(dt.score(val_input, val_target))
````

### Cross validation(교차검증, cross_validate())
> 교차검증: 검증세트를 떼어 내어 평가하는 과정 반복
>
> 직접 검증세트를 떼어내지 않고 훈련세트 전체를 cross_validate() 함수에 전달
>
> 검증&훈련은 DL에서 쓰고,교차는 ML에서 사용


````python
from sklearn.model_selection import cross_validate
scores = cross_validate(dt, train_input, train_target) # default는 5-fold, cv매개변수로 변경 가능
print(scores) # fit_time, score_time, test_score 키를 가진 dictionary를 반환

import numpy as np
print(np.mean(scores['test_score']))  # test_score(5개) 점수의 평균
````

cross_validate()는 훈련 세트를 섞어 폴드를 나누지 않으나, 교차검증을 할 때 훈련세트를 섞으려면 분할기(splifier)를 지정해야함

*현재는 train_test_split()함수로 훈련세트를 준비해서 섞을 필요 없음*

분할기는 교차검증에서 폴드를 어떻게 나눌 지 결정,

cross_validate() 함수는 기본적으로 **회귀모델일 경우 KFold** 분할기, **분류모델 일 때는 stratifiedKFold**를 사용




````python
from sklearn.model_selection import StratifiedKFold
scores = cross_validate(dt, train_input, train_target, cv = StratifiedKFold()) # 객체(dt)에 따라 KFold, Stratified를 자동으로 구분
print(np.mean(scores['test_score']))
````

````python
# K-fold의 k는 n_splits를 통해 지정
splitter = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 42)
scores = cross_validate(dt, train_input, train_target, cv = splitter)
print(np.mean(scores['test_score']))
````

### 하이퍼파라미터 튜닝
Decision tree model에서 max_depth을 고정하고, min_sample_split을 바꿔가며 최적의 값을 찾음

> 한 매개변수의 최적값을 찾고 다른 매개변수의 최적값을 찾을 수 없음..
>
> max_depth의 최적값은 min_samples_split 매개변수의 값이 바뀌면 함께 달라짐

파이썬 for반복문을 통해 구현할 수 있지만 **그리드 서치(Grid Search, GridSearchCV 클래스)** 사용


````python
from sklearn.model_selection import GridSearchCV
# min_impurity_decrease 매개변수의 최적값을 찾기
params = {'min_impurity_decrease': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]}

gs = GridSearchCV(DecisionTreeClassifier(random_state =42), params, n_jobs = -1)
# n_jobs = -1 : 가능한 모든 core를 이용(동시에 여러 모델 훈련)
gs.fit(train_input, train_target)
````

````python
dt = gs.best_estimator_
print(dt.score(train_input, train_target))
print(gs.best_params_) # gridSearch로 찾은 최적의 매개변수
print("최적의 매개변수: " + str(gs.best_params_))

print("평균 점수: ",gs.cv_results_['mean_test_score']) # cv_results_ 속성의 mean_test_score 키에 저장됨
````

첫번째(0.0001) 값이 가장 큼!

argmax()함수를 사용하면 가장 큰 값의 인덱스를 추출할 수 있음

````python
best_index = np.argmax(gs.cv_results_['mean_test_score'])
print(gs.cv_results_['params'][best_index])
````

````python
params = {'min_impurity_decrease': np.arange(0.0001, 0.001, 0.0001),
          'max_depth': range(5, 20, 1),
          'min_samples_split': range(2, 100, 10)}

gs = GridSearchCV(DecisionTreeClassifier(random_state =42), params, n_jobs = -1)
gs.fit(train_input, train_target)
print(gs.best_params_)
print(np.max(gs.cv_results_['mean_test_score']))
````
